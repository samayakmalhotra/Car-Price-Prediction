```python
mask = df.dtypes == np.float #Storing columns having datatype of float
float_cols = df.columns[mask]

skew_limit = 0.75 # define a limit above which we will log transform
skew_vals = df[float_cols].skew()
skew_vals
```

```python
skew_cols = (skew_vals.sort_values(ascending=False).to_frame().rename(columns={0:'Skew'}).query('abs(Skew) > {}'.format(skew_limit)))
skew_cols #Only displays columns with skewness >0.75
```

```python
for col in skew_cols.index.values:
    if col == "price":
        continue
    df[col] = df[col].apply(np.log1p) #Convert the skewed columns to Normal distibition
```

```python
X = df.drop(['wheelbase','carheight'],axis=1) #For our model correlation is not good so we remove the columns having less impact on predicted column. Also helps reducing noise
```

```python
X = X.drop('price',axis=1) #Adding independent features in X
y = df['price'] #Target/Label Column
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import GridSearchCV, train_test_split, KFold
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
#Splitting into test and training
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=72018)
```

```python
#Baseline model for regression
#Mean price for a car
y_pred= y_train.mean()
y_pred
```

```python
y_pred4=y_test.mean()
y_pred4
```

```python
from random import seed
from random import randrange

# zero rule algorithm for regression
def zero_rule_algorithm_regression(y_pred, y_pred4):
	output_values = [row[-1] for row in y_pred]
	prediction = sum(output_values) / float(len(output_values))
	predicted = [prediction for i in range(len(y_pred4))]
	return predicted
```

```python
y_pred2=[y_pred]*len(y_train)
```

```python
#Mean Absolute Error for training data baseline
mean_absolute_error= mean_absolute_error(y_train,y_pred2)
mean_absolute_error
```

```python
#Mean Squared Error for training data baseline
mean_squared_error= mean_squared_error(y_train,y_pred2)
mean_squared_error
```

```python
#Prediction for y test and MSE,MAE
from sklearn import metrics

y_pred3=[y_pred4] * len(y_test)
mae=mean_absolute_error(y_test,y_pred3)
mse=mean_squared_error(y_test,y_pred3)

print(f"MAE:{mae} \nmse: {mse}")
```

```python
#Standardizing the data
s = StandardScaler()
X_train_s = s.fit_transform(X_train)
X_test_s = s.transform(X_test)
```

```python
#Linear Regression
lr = LinearRegression()
lr = lr.fit(X_train_s,y_train)
pred = lr.predict(X_test_s)
r2_score(y_test,pred) #Accuracy score
```
